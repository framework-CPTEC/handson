```python
import metview as mv

def build_layout():
    
    #PageDescription
    
    _Geographical_View_ = mv.geoview(
        map_projection="cylindrical",
        map_area_definition="corners",
        area=[-52.8595, -83.045, 13.982, -20.0063]
        )
    
    _Page_ = mv.plot_page(
        top=0,
        bottom=100,
        left=0,
        right=100,
        view=_Geographical_View_
        )
    
    # SuperPage description
    
    dw = mv.plot_superpage(
        layout_size="A4",
        layout_orientation="LANDSCAPE",
        custom_width=29.7,
        custom_height=21.0,
        pages=[_Page_]
        )
    return dw
```

```python
wrf_url = 'https://dataserver.cptec.inpe.br/dataserver_modelos/wrf/ams_07km/brutos/2025/11/13/00/WRF_cpt_07KM_2025111300_2025111301.grib2'
```

```python
fs = mv.download(url=wrf_url)
```

```python
PSNM = mv.read(
    paramId=260074,
    data=fs
    )
PSNM.values().mean()
```

```python
mv.plot(fs['prmsl'])
```

```python
contour = mv.mcont(
    legend="ON",
    contour="ON",
    contour_shade="ON",
    contour_shade_method="AREA_FILL",
    contour_shade_max_level_colour="RED",
    contour_shade_min_level_colour="BLUE",
    contour_shade_colour_direction="CLOCKWISE"
    )

contour2 = mv.mcont(
    legend="OFF",
    contour="ON",
    contour_shade="OFF",
    contour_shade_max_level_colour="BLACK",
    contour_shade_min_level_colour="BLACK",
    )
```

```python
mv.plot(PSNM, contour,
        PSNM, contour2)
```

```python
area = [-60,-120,20,0] # S,W,N,E
AMS = mv.geoview(
    map_area_definition = "corners",
    area                = area,
    coastlines = mv.mcoast(
        map_coastline_land_shade        = "on",
        map_coastline_land_shade_colour = "#eeeeee",
        map_grid_latitude_increment     = 10,
        map_grid_longitude_increment    = 10)
    )
```

```python
mv.plot(AMS,
        PSNM, contour)
```

```python
text = mv.mtext(
    text_line_1 = "Pressão Reduzida ao Nível do Mar - ANL: 2025/02/19 00Z",
    text_font_size = 0.8,
    text_colour = "black"
)
axis = mv.maxis(
    axis_title_height = 1,
    axis_tick_label_height = 2.
)
dw = build_layout()

mv.plot(dw[0], 
        text, 
        axis,
        PSNM, contour)
```
